## Arquitectura general del modelo de aprendizaje ##

Específicamente esto es un modelo de **"Regresión Logística"** utilizando una estructura de **red neuronal**

Este es el detalle:

<img src="images/LogReg_kiank.jpg" style="width:650px;height:400px;">

**Lo mismo pero explicado matemáticamente**:

Para cada imagen $x^{(i)}$:
$$z^{(i)} = w^T x^{(i)} + b \tag{1}$$
$$\hat{y}^{(i)} = a^{(i)} = sigmoid(z^{(i)})\tag{2}$$ 

Para la función de "pérdida" usamos "entropía cruzada"

$$ \mathcal{L}(a^{(i)}, y^{(i)}) =  - y^{(i)}  \log(a^{(i)}) - (1-y^{(i)} )  \log(1-a^{(i)})\tag{3}$$  

Posteriormente el **costo** es calculado sumando y promediando todas las imágenes usadas en el entrenamiento:
$$ J = \frac{1}{m} \sum_{i=1}^m \mathcal{L}(a^{(i)}, y^{(i)})\tag{6}$$



## Construyendo nuestra red neuronal ## 


### Función de activación


````python
# sigmoid

def sigmoid(z):
    s = 1 / (1 + np.exp(-z))
  
    return s
````

### Inicializando matrices en cero


````python
def initialize_with_zeros(dim):
    w = np.zeros(shape=(dim, 1))
    b = 0

    assert(w.shape == (dim, 1))
    assert(isinstance(b, float) or isinstance(b, int))
    
    return w, b
````

### Propagación y Retro-Propagación

Cuando hablamos de **propagación** hablamos de calcular la salida de cada perceptrón y aplicarle la función de activación.

Cuando hablamos de **retro-propagación** hablamos de, en base a la función de costo obtener valores para posteriormente modificar los pesos y zesgos usados para que en una futura **propagación** podamos obtener valores que disminuyan el costo previamente evaluado.

Al proceso de repetir varias veces la ** propagación y retro-propagación ** se lo llama **ENTRENAMIENTO (u optimización)** 


Propagación:
- dado X
- calcular (perceptrón + activación) $A = \sigma(w^T X + b) = (a^{(0)}, a^{(1)}, ..., a^{(m-1)}, a^{(m)})$
- calcular el costo: $J = -\frac{1}{m}\sum_{i=1}^{m}y^{(i)}\log(a^{(i)})+(1-y^{(i)})\log(1-a^{(i)})$

Derivadas que se usarán en la retro-propagación:

$$ \frac{\partial J}{\partial w} = \frac{1}{m}X(A-Y)^T\tag{7}$$
$$ \frac{\partial J}{\partial b} = \frac{1}{m} \sum_{i=1}^m (a^{(i)}-y^{(i)})\tag{8}$$

````python
def propagate(w, b, X, Y):
    m = X.shape[1]  #cantidad de elementos (fotos) a entrenar
    
    # Propagación con todos los elementos del conjunto de entrenamiento
    A = sigmoid(np.dot(w.T, X) + b)  # perceptrón y activación
    cost = (- 1 / m) * np.sum(Y * np.log(A) + (1 - Y) * (np.log(1 - A)))  # costo
    
    
    # Retro-propagación para encontrar las derivadas
    dw = (1 / m) * np.dot(X, (A - Y).T)
    db = (1 / m) * np.sum(A - Y)

    assert(dw.shape == w.shape)
    assert(db.dtype == float)
    cost = np.squeeze(cost)
    assert(cost.shape == ())
    
    grads = {"dw": dw,
             "db": db}
    
    return grads, cost
````

### Optimización



````python
def optimize(w, b, X, Y, num_iterations, learning_rate, print_cost = False):
    costs = []
    
    # Repetirmos el proceso tantas veces como lo indicamos
    for i in range(num_iterations):
        
        
        # propagamos y retropropagamos
        grads, cost = propagate(w, b, X, Y)
        
        # obtenemos los valores de la derivada
        dw = grads["dw"]
        db = grads["db"]
        
        # actualizamos los pesos y zesgos (teniendo en cuenta la "taza de entrenamiento")
        w = w - learning_rate * dw  
        b = b - learning_rate * db
                
        # pavada: cada 100 iteraciones registramos el costo calculado
        # y si fué indicado por parámetro lo imprimimos.
        if i % 100 == 0:
            costs.append(cost)
        
            if print_cost:
                print ("Costo en la iteración %i: %f" % (i, cost))
    
    params = {"w": w,
              "b": b}
    
    grads = {"dw": dw,
             "db": db}
    
    return params, grads, costs
````

En la función anterior retornamos los pesos y zesgos "aprendidos", por lo tanto con ellos podemos usarlos para analizar cualquier otra imagen que no sea la del conjunto de entrenamiento. 

Creamos la siguiente función para eso:

````python
def predict(w, b, X):
    
    m = X.shape[1] #cantidad de elementos a probar
    Y_prediction = np.zeros((1, m)) #inicializamos en cero la matriz de resultado
    w = w.reshape(X.shape[0], 1) #nos aseguramos que el vector de pesos tenga la forma correcta
    
    # Esto ya lo conocemos: calculamos el perceptrón y le aplicamos la función de activación
    A = sigmoid(np.dot(w.T, X) + b)
    
    #para cada imagen procesada, a su resultado lo normalizamos a 0 o 1
    for i in range(A.shape[1]):
        Y_prediction[0, i] = 1 if A[0, i] > 0.5 else 0
    
    assert(Y_prediction.shape == (1, m))
    
    return Y_prediction
````

## Probando que todo ande y no explote ##

Vamos a hacer todo junto:

Dado:
- los conjuntos de entrenamiento y test,
- el número de repeticiones para el entrenamiento
- la taza de aprendizaje
- y si queremos o no loguear el avance del entrenamiento

hacer:
- inicializar las matrices
- propagar, retropropagar y re-calcular pesos y zesgos
- evaluar el conjunto de entrenamiento y el de test (pa'saber nomá)


````python
def model(X_train, Y_train, X_test, Y_test, num_iterations=2000, learning_rate=0.5, print_cost=False):
    

    # inicializamos pesos y zesgos
    w, b = initialize_with_zeros(X_train.shape[0])

    # optimizamos!!! (propagación ---> retropropagación ---> cálculo de nuevos pesos y zesgos)
    parameters, grads, costs = optimize(w, b, X_train, Y_train, num_iterations, learning_rate, print_cost)
    
    # Obtenemos los pesos y zesgos aprendidos
    w = parameters["w"]
    b = parameters["b"]
    
    # y probamos ambos conjuntos (entrenamiento y test)
    Y_prediction_test = predict(w, b, X_test)
    Y_prediction_train = predict(w, b, X_train)


    # Print train/test Errors
    print("Precisión en entrenamiento: {} %".format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print("Precisión en test: {} %".format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))

    
    d = {"costs": costs,
         "Y_prediction_test": Y_prediction_test, 
         "Y_prediction_train" : Y_prediction_train, 
         "w" : w, 
         "b" : b,
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}
    
    return d
````

Con la siguiente línea hacemos pelota todo..... o no

````python
d = model(train_set_x, train_set_y, test_set_x, test_set_y, num_iterations = 2000, learning_rate = 0.005, print_cost = True)
````

Veamos la evolución del entrenamiento :-) 

````python
costs = np.squeeze(d['costs'])
plt.plot(costs)
plt.ylabel('Costo')
plt.xlabel('Iteraciones (por cada 100)')
plt.title("Taza de entrenamiento: " + str(d["learning_rate"]))
plt.show()
````

## SOLO SI HAY TIEMPO!!!!!  ##



#### ¿qué sucede si cambiamos la taza de entrenamiento? ####

La taza de entrenamiento determina que tan rápido se alteran nuestros parámetros (peso y zesgo por el momento) para alcanzar el objetivo deseado, pero una taza de entrenamiento incorrecta puede resultar en un entrenamiento muuuuuuy lento pero eficaz ("convergencia lenta") o en un entrenamiento muy errático ("overshoot").


````python
learning_rates = [0.01, 0.001, 0.0001, 0.005]
models = {}
for i in learning_rates:
    print ("La taza de entrenamiento es: " + str(i))
    models[str(i)] = model(train_set_x, train_set_y, test_set_x, test_set_y, num_iterations = 1500, learning_rate = i, print_cost = False)
    print ('\n' + "-------------------------------------------------------" + '\n')

for i in learning_rates:
    plt.plot(np.squeeze(models[str(i)]["costs"]), label= str(models[str(i)]["learning_rate"]))

plt.ylabel('costo')
plt.xlabel('iteraciones (por cientos)')

legend = plt.legend(loc='upper center', shadow=True)
frame = legend.get_frame()
frame.set_facecolor('0.90')
plt.show()
````
